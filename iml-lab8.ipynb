```python
'''Practical No.8: Write a Pandas program to create a line plot of
the opening, closing stock prices of given
company between two specific dates.

Name: Tank Yash
Eno: 216270307073
'''

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt

data = pd.read_csv(r'../input/nifty50-stock-market-data/ADANIPORTS.csv')

data['Date'] = pd.to_datetime(data['Date'])

start_dt = '2007-11-27'
end_dt = '2008-02-06'

filtered_data = data[(data['Date'] >= start_dt) & (data['Date'] <= end_dt)]

plt.figure(figsize=(10,7))
plt.plot(filtered_data['Date'], filtered_data['Open'], label='Opening Price', color='red', marker='o')
plt.plot(filtered_data['Date'], filtered_data['Close'], label='Closing Price', color='green', marker='x')

plt.title('Opening and Closing Stock Prices of ADANIPORTS')
plt.xlabel('Date')
plt.ylabel('Price')
plt.xticks(rotation=50)
plt.legend()

plt.tight_layout()
plt.show()
```
